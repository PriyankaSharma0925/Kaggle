# In this kernel we will try to extract data via API.
The data we get via APIs is in structured format.Its easy to use. 
To extract data from API we need API endpoint and API key.

**With the help of request library we will hit the server with the below url.**

```python
url="https://api.themoviedb.org/3/movie/top_rated?api_key=5573865962d8153adc3efb31b4e7e5c4&language=en-US&page=1"
```

**With the help of get method we will take the response of this page into a variable *page***

```python
page=requests.get(url)
```

**To look at the response in json format, we will use the json method()**

```python
page.json()
```

**Above we got the data of the first page and from the output we can see there are a total of 498 pages.**
Here we are interested in results part only. So now we will extract the result part.

```python
result=page.json()['results']
result
```

# Now we will get this data in a pandas dataframe.

```python
df=pd.DataFrame(result)
```

```python
df
```

# To get the data from all the pages we will run a for loop.

```python
total_pages=page.json()['total_pages']

total_pages
```

```python
dataFrame=pd.DataFrame()
for i in range(1,total_pages+1):
    page=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=5573865962d8153adc3efb31b4e7e5c4&language=en-US&page={}".format(i))
    #print(page.json())
    result=page.json()['results']
    temp=pd.DataFrame(result)
    dataFrame.append(temp,ignore_index=True)
```
